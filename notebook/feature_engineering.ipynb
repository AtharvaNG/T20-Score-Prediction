```python
import pandas as pd
import os
import numpy as np
```

```python
df=pd.read_pickle("data/ball_df.pkl")
```

```python
type(df)
```

```python
df['batting_team'].unique()
```

```python
valid_teams=['Pakistan',
 'India',
 'New Zealand',
 'South Africa',
 'West Indies',
 'Sri Lanka',
 'Bangladesh',
 'Zimbabwe',
 'England',
 'Australia']
```

# Select only specific teams

```python
df=df[
    df['batting_team'].isin(valid_teams) &
    df['bowling_team'].isin(valid_teams)
].copy()
```

```python
df.head()
```

```python
df.shape
```

# Sort the data by match_id and ball
because we need to calculate cumilative sum and other things which will work only if the data is sorted

Group rows by match, and within each match, order the balls in sequence

```python
df=df.sort_values(
    by=['match_id','ball']
).reset_index(drop=True)
```

```python
df
```

# Processing city
there are many rows with city as null so for that particular rows we will extract the city from the venue
And also there are many cities where only 1 match is played so we will be removing that match from the data for better recommendations.
We will consider only the cities where at least 5 matches are played so approx 5x120=600 balls are bowled

```python
df.isnull().sum()
```

```python
df[df['city'].isnull()]['venue'].value_counts()
```

```python
cities=np.where(df['city'].isnull(),df['venue'].str.split().apply(lambda x:x[0]),df['city'])
```

```python
df['city']=cities
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['venue'],inplace=True)
```

```python
df.head()
```

```python
df['city'].value_counts()
```

```python
city_count=df['city'].value_counts()
city_count=city_count[city_count>599]
valid_cities=city_count.index.tolist()

df=df[df['city'].isin(valid_cities)].copy()
df
```

```python
df['city'].unique()
```

# Calculate current runs(cummelative runs)

```python
df["current_score"]=(
    df.groupby("match_id")["runs"].cumsum()
)
```

```python
df.head()
```

# Calculate wickets fallen and wickets left

```python
df['wickets_fallen']=(
    df.groupby('match_id')['wicket'].cumsum()
)
```

```python
df['wickets_left']=10-df['wickets_fallen']
```

```python
df.head()
```

# Calculating balls left

```python
df['ball_number']=(
    df['ball'].astype(str).str.split(".").str[0].astype(int)*6+
    df['ball'].astype(str).str.split(".").str[1].astype(int)
)
```

```python
df['balls_left']=120-df['ball_number']
df['balls_left']=df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

# Calculating current run rate

```python
df['crr']=df['current_score']/(df['ball_number']/6)
```

```python
df
```

# Calculating last 5 overs runs

```python
df['last_five_overs_runs']=(
    df.groupby('match_id')['runs']
    .rolling(window=30)
    .sum()
    .reset_index(level=0,drop=True)
)
```

```python
df.head(50)
```

# Calculating final score

```python
df['final_score']=(
    df.groupby('match_id')['runs']
    .transform('sum')
)
df
```

# Droping null values for last 5 overs

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','crr','last_five_overs_runs','final_score']]
```

```python
df.reset_index(drop=True)
```

# Shuffling the data to avoid baised predictions

```python
df=df.sample(df.shape[0]).reset_index(drop=True)
```
